```python
from google.colab import drive
from Bio import SeqIO
import pandas as pd
import numpy as np
import re
```

# Load Data

```python
file_path = '/content/gdrive/MyDrive/Colab Notebooks/Dataset/DM5.fas'
drive.mount('/content/gdrive')

with open(file_path, 'r') as f:
    file_str = f.read()

# Split the file string into individual sequence records
records = file_str.split('>')

# Remove empty strings
records = [rec for rec in records if len(rec) > 0]

records
```

```python
# Split each record into its ID and sequence
df_list = []
for rec in records:
    rec_split = rec.split('\n')
    seq_id = rec_split[0]
    sequence = ''.join(rec_split[1:])
    df_list.append({'id': seq_id, 'sequence': sequence})

# Create a Pandas DataFrame from the list of sequence records
df = pd.DataFrame(df_list)

df
```

```python
new_column = pd.Series([1] * 410)

df['class'] = new_column

df = df.drop('id', axis=1)
```

```python
df
```

## Cek selain A, C, G, T

```python
invalid_chars = []
for sequence in df['sequence']:
    invalid = ''.join(set(sequence.upper()) - set('ACGT'))
    if invalid:
        invalid_chars.append(invalid)

# Display the erroneous characters or punctuation marks
if invalid_chars:
    print("Invalid characters or punctuation marks found:")
    for invalid in invalid_chars:
        print(invalid)
else:
    print("No invalid characters or punctuation marks found.")
```

# Preprocessing

## Bersihkan selain A, C, G, T

```python
# Function to preprocess DNA sequence
def preprocess_sequence(sequence):
    # Remove invalid characters
    processed_sequence = re.sub('[^ACGT]', '', sequence)
    return processed_sequence

# Preprocess the 'Sequence' column
df['sequence'] = df['sequence'].apply(preprocess_sequence)

# Print the updated dataframe
df
```
